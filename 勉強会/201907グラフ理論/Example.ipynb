```python
### Random Graph

def Edges_Random_Graph(start: int,end: int, p: float, directed=False):
    
    output=[]
    if directed:
        for i in range(start,end+1):
            for j in range(start,end+1):
                r=np.random.random()
                if r<=p:
                    output.append((i,j))
    else:
        for i in range(start,end):
            for j in range(i+1,end+1):            
                r=np.random.random()
                if r<=p:
                    output.append((i,j))
    return output                

edge_list=Edges_Random_Graph(1,20,0.7)

```

```python

### add nodes or edges to a graph

G.add_node('a node')
G.add_edge('A','B')

node_list=[100,101,102]
edge_list=[(1,100),(2,102)]
G.add_nodes_from(node_list)
G.add_edges_from(edge_list)


nx.draw_networkx(G)
plt.show()

###
```

```python
temp=Edges_Random_Graph(10,30,0.2)
edge_list.extend(temp)

temp=Edges_Random_Graph(20,40,0.7)
edge_list.extend(temp)

G=nx.Graph()
G.add_edges_from(edge_list)

nx.draw_networkx(G)
plt.show()
```

```python
### access the information of a graph

# info of nodes and graph
for node in G.nodes():
    print(G[node])
for edge in G.edges():
    print(edge,G[edge[0]][edge[1]])

```

```python
# degrees
if not nx.is_directed(G):
    # 無向グラフの場合
    for node in G.nodes():
        print(node,G.degree(node))
else:
    # 有向グラフの場合
    for node in G.nodes:
        print(node,'  in',G.in_degree(node),'  out',G.out_degree(node))
```

```python

### 辞書の中身を表示するヘルパー

def dict_show(dictionary):
    for key in dictionary.keys(): 
        print(key,':',dictionary[key])
        
### centrality

# degree centrality        
if not nx.is_directed(G):
    deg=nx.degree_centrality(G)
    print('Degree Centrality')
    dict_show(deg)
else:
    in_deg=nx.in_degree_centrality(G)
    print('In Degree Centrality')
    dict_show(in_deg)
    
    out_deg=nx.out_degree_centrality(G)
    print('Out Degree Centrality')
    dict_show(out_deg)
    
```

```python

# betweenness centrality
bet=nx.betweenness_centrality(G)
print('Betweenness Centrality')
dict_show(bet)
```

```python

# PageRank
PageRank=nx.pagerank_numpy(G)   
print('PageRank')
dict_show(PageRank)

###

```

```python
### コミュニティ検出

communities=community.best_partition(G)
dict_show(communities)

colors='bgrcmykw'
colors=list(colors)
color_map=[]
for node in G.nodes:
    color_map.append(colors[communities[node]])
nx.draw_networkx(G,node_color=color_map)
plt.show()
```
